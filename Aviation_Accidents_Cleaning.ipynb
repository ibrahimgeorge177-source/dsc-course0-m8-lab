### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
#cleaning injury columns

inj_columns = [
    'Total.Fatal.Injuries',
    'Total.Serious.Injuries',
    'Total.Minor.Injuries',
    'Total.Uninjured'
]

df1[inj_columns] = df1[inj_columns].fillna(0)
df1[inj_columns]
```

### Observation


- The injury-related columns (Total.Fatal.Injuries, Total.Serious.Injuries, Total.Minor.Injuries, Total.Uninjured) were successfully cleaned, resulting in complete numerical values across all records.

- The dataset now contains consistent injury data with no missing values, enabling accurate calculation of total and proportional injury rates.

- Most records show zero or low injury counts, indicating that the majority of aviation accidents in the dataset are non-fatal or involve minimal harm.

- The cleaned data preserves both severe and non-severe outcomes, ensuring that subsequent safety metrics are based on complete and reliable information.

```python
# creating total passengers

df1['Total_Passengers_Est'] = (
    df1['Total.Fatal.Injuries'] +
    df1['Total.Serious.Injuries'] +
    df1['Total.Minor.Injuries'] +
    df1['Total.Uninjured']
)

df1
```

### Observations

- The Total_Passengers_Est column was successfully created by summing fatal, serious, minor, and uninjured counts for each record.

- This derived field provides an estimate of the total number of occupants involved in each aviation incident, since the dataset does not directly provide passenger counts.

- The values vary significantly across records, ranging from small aircraft incidents with few occupants to large-scale events involving hundreds of individuals.

- This engineered feature enables normalization of injury outcomes, allowing for meaningful comparison of safety risks across different aircraft types and flight sizes.

```python
# Filter out records with zero passengers to avoid division by zero errors
df1 = df1[df1['Total_Passengers_Est'] > 0]
df1
```

### Observations

- The dataset was filtered to remove records where Total_Passengers_Est equals zero to prevent invalid division operations during rate calculations.

- This step ensures that all computed injury rates (fatal and serious injury rates) are mathematically valid and free from infinite or undefined values.

- Removing zero-passenger records improves data integrity and ensures that safety metrics are based on meaningful flight occupancy values.

- This cleaning step helps maintain consistency in proportional analysis across all aircraft models by ensuring comparable denominators.

```python
#safety metrics fatal injury rate

df1['Fatal_Injury_Rate'] = df1['Total.Fatal.Injuries'] / df1['Total_Passengers_Est']
df1['Fatal_Injury_Rate']
```

### Observations

- The Fatal_Injury_Rate column represents the proportion of passengers who died in each accident relative to the estimated total number of occupants.

- The values range from 0.0 to 1.0, indicating that some accidents resulted in no fatalities while others resulted in all occupants being fatally injured.

- Most values are 0.0, suggesting that the majority of aviation accidents in the dataset do not result in fatalities.

- Occasional extreme values (e.g., 1.0) indicate rare but highly severe accidents where all occupants were fatally injured.

```python
#serious injury rate

df1['Serious_Injury_Rate'] = df1['Total.Serious.Injuries'] / df1['Total_Passengers_Est']
df1['Serious_Injury_Rate']
```

### Observations

- The Serious_Injury_Rate column represents the proportion of passengers who sustained serious injuries in each accident relative to the estimated total onboard.

- The values range between 0.0 and 0.5 in the displayed sample, indicating that serious injuries are generally limited and do not typically affect all occupants.

- A large number of records have a value of 0.0, suggesting that most aviation accidents do not result in serious injuries.

- Occasional higher values (e.g., 0.5) indicate cases where a significant portion of passengers were seriously injured, highlighting variability in accident severity.

```python
#total severe risk

df1['Severe_Injury_Rate'] = (
    df1['Total.Fatal.Injuries'] + df1['Total.Serious.Injuries']
) / df1['Total_Passengers_Est']

df1['Severe_Injury_Rate']
```

### Observations

- The Severe_Injury_Rate column was successfully created, representing the proportion of passengers who suffered either fatal or serious injuries per accident.

- The values range from 0.0 to 1.0, indicating that some accidents resulted in no severe injuries, while others resulted in all occupants being severely injured.

- A large number of records have a value of 0.0, suggesting that most aviation accidents in the dataset do not result in fatal or serious injuries.

- The presence of intermediate values (e.g., 0.5) indicates mixed severity outcomes, where only a portion of occupants experienced severe injuries.
